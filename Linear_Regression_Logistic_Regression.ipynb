```python
class linear_regression():

  def __init__(self, learning_rate = 0.1, iterations = 200):
    self.learning_rate = learning_rate
    self.iterations = iterations

  def update_weights(self):
    # Predict using training data
    y_pred = self.x.dot(self.w) + self.b

    # calculating gradient descent
    dw = (-2 * (self.x.T).dot(self.y - y_pred)) / self.m
    db = (-2 * np.sum(self.y - y_pred)) / self.m

    # Update weights using gradient descent
    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db
    return self

  def fit(self, x, y):
    self.m, self.d = x.shape
    self.w = np.zeros(self.d)
    self.x = x
    self.y = y
    self.b = 0

    self.error_list = []

    for i in range(self.iterations):
      self.update_weights()

    return self

  def predict(self, X):
    return X.dot(self.w) + self.b

  def r2_score(self,x,y):
    y_ = self.predict(x)
    ss_pred = np.sum((y-y_)**2)
    ss_mean = np.sum((y.mean() - y)**2)
    rscore = 1 - ss_pred/ss_mean
    return rscore
```

```python
class LogisticRegression():
  def __init__(self , learning_rate = 0.01, iterations = 200):
    self.iterations = iterations
    self.learning_rate = learning_rate

  def update_weights(self):
    y_ = 1/(1 + np.exp(-1*(self.x.dot(self.w) + self.b)))
    dw = self.x.T.dot(y_ - self.y)/self.m
    db = (np.sum(y_ - self.y)) / self.m

    self.w = self.w - self.learning_rate * dw
    self.b = self.b - self.learning_rate * db

  def fit(self , x ,y):
    self.m , self.d = x.shape
    self.w = np.zeros(self.d)
    self.b = 0
    self.x = x
    self.y = y

    for i in range(self.iterations):
      self.update_weights()
  def predict(self, x):
    output = []
    for i in range(len(x)):
        prob = 1 / (1 + np.exp(-(x[i].dot(self.w) + self.b)))
        pred = 1 if prob > 0.5 else 0
        output.append([pred, prob])
    return output
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = data.data
Y = data.target

# Normalize (optional but helpful)
X = (X - X.mean(axis=0)) / X.std(axis=0)
```
